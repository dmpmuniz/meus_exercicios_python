```python
import requests

# Set the API endpoint URL
url = "https://restcountries.com/v3.1/all"
```

```python
# Send a GET request to the REST Countries API
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
```

```python
def pais():
    for i in data:
        nome = i["name"]["official"]
        popu = i["population"]
        try:
            dic_moeda = i["currencies"]
        except:
            print("O Pais:",nome,"Tem a população estinada em:",popu,"Habitantes","Este Pais não possui moeda corrente. ") 
            print("")
        lis_moeda = dic_moeda.keys()
        for l in lis_moeda:
            moeda = l
            print("O Pais:",nome,"Tem a população estinada em:",popu,"Habitantes","sua moéda é:",moeda,"ou", dic_moeda[moeda]["name"] )
            print("") 
```

```python
pais()
```
